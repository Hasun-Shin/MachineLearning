What was the average (mean) BasePay of all employees per year? (2011-2014) ?

```python
year = (df['Year']>=2011) & (df['Year']<=2014)

year_df = df[year]

year_df['BasePay'].groupby(year_df['Year']).mean()
```

How many people have the word Chief in thier job title? (This is pretty tricky) (using lambda expression)

```python
df_jobs = df['JobTitle']

val = df_jobs.values

count = 0
for i in val :
    if 'Chief' in i:
        count = count +1
print(count)
```
