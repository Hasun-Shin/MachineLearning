```python
import numpy as np
from sklearn import datasets, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
# from keras.models import Sequential #그리드 서치 하면 안 써도 됨. 
from keras.layers.core import Dense, Activation #그리드 서치하면 안 써도 됨. 

from keras.utils import np_utils
from keras import backend as K
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam

from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, GlobalMaxPooling2D
from tensorflow.keras.models import Model
```

```python
iris = datasets.load_iris()
x = preprocessing.scale(iris.data)
y = np_utils.to_categorical(iris.target)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size  = 0.7)
num_classes = y_test.shape[1] # 3 setosa, vesicolor, virginica
```

```python
num_classes
```

```python
x_train.shape
```

```python
# model = Sequential()

def iris_model(activation='relu', optimizer='adam' ,out_dim ='100'):
#     model = Sequential()
#     model.add(Dense(out_dim, input_dim=4, activation=activation)) #입력데이터가 input_dim 꽃잎, 꽃받침 등등. 
#     model.add(Dense(out_dim, activation=activation))
#     model.add(Dense(3, activation='softmax'))
    
    i = Input(4)
    x = Dense(out_dim, activation=activation)(i)
    x = Dense(out_dim, activation=activation)(x)
    x = Dense(num_classes, activation='softmax')(x)
 
    
    
    model=Model(i,x) # 각 조합마다 테이블 프린트됨. 
    
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model


```

```python
#train
#옵션값은 더 늘려도 된다. 시간이 된다면.. 
activation = ["relu", "sigmoid"]
optimizer = ["adam","adagrad"]
out_dim = [100, 200]
nb_epochs = [10, 25]
batch_size=[5,10]

#2x2x2x2x2 = 32가지 경우의 수 가 나올 것임. 
```

```python
model = KerasClassifier(build_fn=iris_model, verbose=0)  #verbose = 0 은 epoch 진행 bar를 보이지 않게 함. 
param_grid = dict(activation=activation,
                 optimizer=optimizer,
                 out_dim=out_dim,
                 batch_size=batch_size,
                 nb_epoch=nb_epochs)

grid = GridSearchCV(estimator=model, param_grid=param_grid)
```

```python
grid_result = grid.fit(x_train, y_train)
```

```python
print(grid_result.best_score_)
print(grid_result.best_params_)
#최적화된 스코어가 나옴. 
```

```python
new_model = iris_model(activation=grid_result.best_params_['activation'],
                      optimizer=grid_result.best_params_['optimizer'],
                      out_dim=grid_result.best_params_['out_dim'])
```
