## 머신러닝 알고리즘 선택하기

### 1. 먼저, 데이터를 잘 전처리한다.
### 2. 알고리즘을 적절히 선택한다




```python
# 라이브러리 읽어 들이기 --- (*1)
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# 학습 전용 데이터와 결과 준비하기 --- (*2)
# X , Y
learn_data = [[0,0], [1,0], [0,1], [1,1]]
# X and Y
learn_label = [0, 0, 0, 1]

# 알고리즘 지정하기(LinierSVC) --- (*3)
clf = LinearSVC()

# 학습전용 데이터와 결과 학습하기  --- (*4) fit, 학습을 시켜주는 함수 
clf.fit(learn_data, learn_label)

# 테스트 데이터로 예측하기 --- (*5) predict 예측 하는 함수. 
test_data = [[0,0], [1,0], [0,1], [1,1]]
test_label = clf.predict(test_data)

# 예측 결과 평가하기 --- (*6)
print(test_data , "의 예측 결과: " ,  test_label)
print("정답률 = " , accuracy_score([0, 0, 0, 1], test_label))
```

```python
# 학습 전용 데이터와 결과 준비하기
# X , Y
learn_data = [[0,0], [1,0], [0,1], [1,1]]
# X xor Y
learn_label = [0, 1, 1, 0]  #(*) xor 전용 레이블로 변경

# 알고리즘 지정하기(LinierSVC)
clf = LinearSVC()

# 학습전용 데이터와 결과 학습하기 
clf.fit(learn_data, learn_label)

# 테스트 데이터로 예측하기
test_data = [[0,0], [1,0], [0,1], [1,1]]
test_label = clf.predict(test_data)

# 테스트 결과 평가하기
print(test_data , "의 예측 결과: " ,  test_label)
print("정답률 = " , accuracy_score([0, 1, 1, 0], test_label))  #(*) xor 전용 레이블로 변경

#정답률이 잘 안나오면 데이터를 바꾸거나 알고리즘을 바꾸거나. 
```

```python
#위에 LinearSVC 알고리즘이 결과가 너무 안나와서 KNeighbor(나이브베이지)사용해보자 

from sklearn.neighbors import KNeighborsClassifier

# 알고리즘 지정하기(KNeighborsClassifier) --- (*2)
clf = KNeighborsClassifier(n_neighbors = 1)

# 학습전용 데이터와 결과 학습하기 
clf.fit(learn_data, learn_label)

# 테스트 데이터로 예측하기
test_data = [[0,0], [1,0], [0,1], [1,1]]
test_label = clf.predict(test_data)

# 테스트 결과 평가하기
print(test_data , "의 예측 결과: " ,  test_label)
print("정답률 = " , accuracy_score([0, 1, 1, 0], test_label))  #(*) xor 전용 레이블로 변경
```
